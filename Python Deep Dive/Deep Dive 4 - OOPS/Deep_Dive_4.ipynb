```python
def fibo(n):
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        seq = fibo(n - 1)
        value = seq[-1] + seq[-2]
        seq.append(value)
        return seq

```

```python
fibo(5)
```

```python
def fibo(n):
    if n < 2:
        seq = [0, 1]
    else:
        seq = fibo(n-1)
        seq.append(seq[-1] + seq[-2])
    
    return seq
```
